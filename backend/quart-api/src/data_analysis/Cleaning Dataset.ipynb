```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import numpy as np

X = specialist_df.iloc[:,1:].drop(columns=['Disease'])
y = disease_vs_symptoms_df['Disease']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

noise_level = 0.2
X_train_noisy = X_train.copy()
total_cells = X_train_noisy.shape[0] * X_train_noisy.shape[1]
num_flips = int(total_cells * noise_level)

row_indices = np.random.choice(X_train_noisy.shape[0], size=num_flips)
col_indices = np.random.choice(X_train_noisy.shape[1], size=num_flips)

for r, c in zip(row_indices, col_indices):
    original_value = X_train_noisy.iloc[r, c]
    # Flip the value (0 -> 1 and 1 -> 0)
    X_train_noisy.iloc[r, c] = 1 - original_value 

print(f"Using noisy features for training. Flips injected: {num_flips}")

# Use the noisy data for training
X_train = X_train_noisy

# Define models with complexity constraints
models = {
    # C=0.1 for stronger regularization
    "Logistic Regression": LogisticRegression(C=0.1, max_iter=1000, random_state=42),
    # max_depth=10 to limit complexity
    "Random Forest": RandomForestClassifier(max_depth=10, random_state=42),
    # C=0.1 for stronger regularization
    "Support Vector Machine": SVC(C=0.1, probability=True, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

trained_models = {}

# Train and evaluate each model
for name, model in models.items():
    
    model.fit(X_train, y_train)
    trained_models[name] = model
    # Evaluation still uses the clean X_test
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Constrained + Noisy {name} Accuracy: {accuracy:.4f}")
```

```python
symptom_columns = list(specialist_df.drop(columns=['Unnamed: 0','Disease']).columns)
symptom_index = {col.lower().replace(" ", "_"): col for col in symptom_columns}

def encode_symptoms(symptom_list):
    row = np.zeros(len(symptom_columns), dtype=int)
    
    for s in symptom_list:
        # Clean the input symptom for lookup
        s_clean = s.lower().strip().replace(" ", "_")
        # Handle the special case found in the dataset: 'dischromic _patches'
        if s_clean == 'dischromic__patches':
            s_clean = 'dischromic _patches'
        
        # Check for cleaned symptom in the index map
        if s_clean in symptom_index:
            col_name = symptom_index[s_clean]
            col_idx = symptom_columns.index(col_name)
            row[col_idx] = 1
        # Handle the case where the key in symptom_index might not be perfectly clean, 
        # but the column name is (e.g., 'dischromic _patches' is a key in symptom_index)
        elif s_clean in symptom_columns:
              col_idx = symptom_columns.index(s_clean)
              row[col_idx] = 1
            
    return row.reshape(1, -1)
```

```python
import numpy as np

# Get the trained Random Forest model
rf_model = trained_models["K-Nearest Neighbors"]

# Predict probabilities on the test set
y_proba = rf_model.predict_proba(X_test)

# Get the class labels (Specialist names)
classes = rf_model.classes_

# Function to get the top N predictions
def get_top_n_predictions(probabilities, classes, n=3):
    # Get the indices that would sort the probabilities in descending order
    top_n_indices = np.argsort(probabilities)[::-1][:n]
    # Get the corresponding classes and probabilities
    top_n_classes = classes[top_n_indices]
    top_n_probabilities = probabilities[top_n_indices]
    
    return list(zip(top_n_classes, top_n_probabilities))

# Get top 3 predictions for the first 5 samples in the test set
top_3_results = []
for i in range(5):
    probs = y_proba[i]
    top_3 = get_top_n_predictions(probs, classes, n=3)
    
    # Get the actual specialist for comparison from y_test
    actual_specialist = dictionary[y_test.iloc[i]]
    
    # Store the result
    top_3_results.append({
        "Actual Specialist": actual_specialist,
        "Top 3 Predictions": top_3
    })

# Display the results
for i, result in enumerate(top_3_results):
    print(f"--- Test Sample {i+1} ---")
    print(f"Actual Specialist: {result['Actual Specialist']}")
    print("Top 3 Predicted Specialists and Probabilities:")
    for specialist, prob in result['Top 3 Predictions']:
        print(f"  - {specialist}: {prob:.4f}")
    print("-" * 20)
```
